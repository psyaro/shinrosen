```python
import pandas as pd
import geopandas as gpd
from settings import geodatadir
from os.path import join
D = 'GeoJSON'
geojson = 'GeoJSON'
```

```python
stations = pd.read_csv(join(geodatadir, '../ekidata/station20220720free.csv'))
stations = stations[['station_cd', 'station_g_cd', 'line_cd', 'station_name', 'lat', 'lon', 'e_sort']]
railways = pd.read_csv(join(geodatadir, '../ekidata/line20220720free.csv'))
railways = railways[['line_cd', 'company_cd', 'line_name', 'line_name_h']]
railway_company = pd.read_csv(join(geodatadir, '../ekidata/company20220401.csv'))
railway_company = railway_company[['company_cd', 'company_name']]
df = pd.merge(railway_company, railways, on='company_cd')
df = pd.merge(df, stations, on='line_cd')
df.to_csv(join(geodatadir, '../ekidata/ekidata.csv'), index=None)
```

```python
df2 = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.lon, df.lat), crs=4326).drop(['lat', 'lon'], 1)
df2.to_file(join(geodatadir, '../ekidata/ekidata.geojson'), driver='GeoJSON')
```

```python
gakkou = gpd.read_file('../data/others/school/P29-21.shp', crs=6668)
gakkou = gakkou[(gakkou.P29_003 >= 16002) & (gakkou.P29_003 <= 16007) & (gakkou.P29_007 == 1)]
gakkou.to_file('gakkou1.geojson', driver='GeoJSON')
```
